```
def extract_desc(html_page):
  desc = html_page.select('div#content article p span')[0].text
  return desc
```

```
def extract_img(html_page):
  img_url = html_page.select('div p img')[0]['src']
  return img_url
```

```
def extract_warning(html_page):
    # find all p tags
    p_tags = html_page.find_all('p')
    # identify starting and ending p tags which correspond to warning and dosage paragraph
    for p in p_tags:
        if 'คำเตือน' in p.text:
            start = p
        if 'ปริมาณการใช้ยา' in p.text:
            end = p
            break
    # find all siblings below warnings
    sibs_down = start.find_next_siblings()
    # find all siblings above dosage
    sibs_up = end.find_previous_siblings()
    # find intersection
    warnings = [x for x in sibs_down if x in sibs_up]
    warnings_list = []
    if len(warnings) == 1:
      warnings_list = [ warning.text for warning in  warnings[0].select('li')]
    elif len(warnings) > 1 :
      for i in warnings:
        if len(i.find_all('li')) != 0:
          for j in i.select('li'):
            warnings_list.append(j.text)
    return warnings_list
```

```
def extract_name(html_page):
  return html_page.select_one('div.entry-content-detail h1').text
```

```
def extract_effect(html_page):
   elements = html_page.select('div.entry-content-detail ul ')
   effect_list = []
   for element in elements:
     effect_list.append(element.text)
   return effect_list
```

```
def get_pobpad_med_data(url):
  response = requests.get(url)
  html_page = bs4.BeautifulSoup(response.content, 'lxml')

  name = extract_name(html_page)
  desc = extract_desc(html_page)
  img = extract_img(html_page)
  warning = extract_warning(html_page)
  infos = extract_effect(html_page)

  med_data = {
      'url': url,
      'name': name,
      'desc': desc,
      'img': img,
      'warning': warning,
      'infos': infos
  }
  return med_data
```

```
from pprint import pprint

med_data = get_pobpad_med_data('https://www.pobpad.com/abacavir-%E0%B8%AD%E0%B8%B0%E0%B8%9A%E0%B8%B2%E0%B8%84%E0%B8%B2%E0%B9%80%E0%B8%A7%E0%B8%B5%E0%B8%A2%E0%B8%A3%E0%B9%8C')

pprint(med_data)
```
